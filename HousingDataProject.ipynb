```python
import pandas as pd

import numpy as np

housing_data = pd.read_csv("/Users/Brytechef/Desktop/AI Engineer Training Program/housing.csv")

housing_data
```

```python
housing_data.dropna(inplace = True)
```

```python
housing_data.info()
```

```python
# To chech the entire columns contained in the dataset
print(housing_data.columns)
```

```python
# The ocean_proximity column, which contains string values, will be converted to categorical variables using pd.get_dummies to facilitate analysis.
# The conversion of text to zeros and ones is referred to as one hot encoding
dummies = pd.get_dummies(housing_data['ocean_proximity']).astype(int)

dummies
```

```python
# The categorical table (dummy variables) will be merged with the original dataset (housing_data) using pd.concat to ensure all features are incorporated for analysis.
housing_data = pd.concat([housing_data, dummies], axis=1)

housing_data
```

```python
# After converting the ocean_proximity column to categorical values, it will subsequently be removed from the dataset for further analysis.

housing_data = housing_data.drop('ocean_proximity', axis = 1)
```

```python
# The dataset will be divided into input features (X) and the target variable (y).
# All columns except median_house_value will serve as input features (X), while median_house_value will be the target variable (y).

x = housing_data.drop(['median_house_value'], axis = 1)

y = housing_data['median_house_value']
```

```python
x
```

```python
y
```

```python
# The dataset will be split into training and testing sets using the train_test_split function from sklearn.model_selection.
# The model will be trained on 80% of the data, with the remaining 20% reserved for testing.

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
# The dataset will be standardized using StandardScaler from sklearn.preprocessing, as the columns do not have specified units.

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)

x_test = scaler.transform(x_test)
```

```python
# The LinearRegression model will be imported from sklearn.linear_model and trained on the dataset.

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model = LinearRegression()

model.fit(x_train, y_train)
```

```python
# To retrieve the intercept and coefficients of the trained linear regression model.
# model.intercept_ gives the y-intercept of the regression equation, while model.coef_ provides the coefficients for each input feature

intercept = model.intercept_
coefficients = model.coef_

print("Linear Regression Equation:")
equation = f"y = {intercept:.2f}"
for i, coef in enumerate(coefficients):
    equation += f" + ({coef:.2f}) * x{i+1}"
print(equation)
```

```python
# To generate predictions for the target variable (y) using the trained linear regression model.

y_pred = model.predict(x_test)
```

```python
# To check the accuracy of the model, we find the mean_squared_error and the r2 score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r2}")
```

```python
# To visualize the performance of the regression model we create a scattered plot from matplotlib.pyplot

import matplotlib.pyplot as plt

# Plot predicted vs actual values
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.show()

# Plot residuals
plt.scatter(y_pred, y_pred - y_test)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residuals vs Predicted Values")
plt.axhline(0, color='red', linestyle='--')
plt.show()
```
